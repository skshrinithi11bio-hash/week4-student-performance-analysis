```python
import pandas as pd
# Create student performance dataset
data = {
    "Student": ["Asha", "Ravi", "Neha", "Arun", "Priya", "Karthik"],
    "Maths": [85, 78, 92, 65, 88, 74],
    "Science": [80, 82, 90, 70, 85, 76],
    "English": [78, 75, 88, 68, 90, 72]
}

df = pd.DataFrame(data)

# Save ONLY CSV to Google Drive
csv_path = "/content/drive/MyDrive/arena files/student_performance.csv"
df.to_csv(csv_path, index=False)

print("CSV saved in Drive at:", csv_path)
df
```

```python
import pandas as pd

# Load CSV from Drive
df = pd.read_csv("/content/drive/MyDrive/arena files/student_performance.csv")

# Calculate total and average marks
df["Total"] = df["Maths"] + df["Science"] + df["English"]
df["Average"] = df["Total"] / 3

# Metrics
class_average = df["Average"].mean()
top_student = df.loc[df["Total"].idxmax(), "Student"]

print("STUDENT PERFORMANCE ANALYSIS")
print("--------------------------------")
print(f"Class Average Marks : {class_average:.2f}")
print(f"Top Performing Student: {top_student}")
```

```python
import matplotlib.pyplot as plt

df.plot(x="Student", y="Average", kind="bar")
plt.title("Average Marks of Students")
plt.ylabel("Average Marks")
plt.show()
```

```python
subjects_avg = df[["Maths", "Science", "English"]].mean()

subjects_avg.plot(marker="o")
plt.title("Average Marks by Subject")
plt.ylabel("Marks")
plt.show()
```
